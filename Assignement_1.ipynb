```python
import torch
import torchvision
from torchvision import transforms
from torchvision.transforms import ToTensor
```

References : 1. USED THIS TUTORIALS TO IMPORT DATA AND USED TA MODULES UPLODED ON CANVAS (https://www.youtube.com/watch?v=ZBhhQ3lZSKY, https://www.youtube.com/watch?v=ovZ_54IUSU4)

```python
train_dataset_path = './drive/MyDrive/Datasets/train'
val_dataset_path = './drive/MyDrive/Datasets/validation'
```

```python
mean = []
std = []

train_transforms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomRotation(10),
    transforms.ToTensor()
    #transforms.Normalize(torch.Tensor(mean),torch.Tensor(std))

])
val_transforms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor()

])
```

```python
train_dataset = torchvision.datasets.ImageFolder(root = train_dataset_path, 
                            transform = train_transforms) 
val_dataset = torchvision.datasets.ImageFolder(root = val_dataset_path, 
                            transform = val_transforms) 
```

```python
batch_size = 50
trainloader = torch.utils.data.DataLoader(train_dataset,
                                          batch_size=batch_size,
                                          shuffle = True,
                                          num_workers=1  
)
valloader = torch.utils.data.DataLoader(val_dataset,
                                          batch_size=batch_size,
                                          shuffle = True,
                                          num_workers=1  
)
valloader
```

```python
import numpy as np
import matplotlib.pyplot as plt

batch= next(iter(trainloader))
images,labels = batch
grid = torchvision.utils.make_grid(images, nrow=10)
plt.figure(figsize=(20,20))
plt.imshow(np.transpose(grid,(1,2,0)))
print('Labels: ',labels)
```
